```python
import numpy as np
import matplotlib.pylab as plt
from scipy.stats import lognorm
```

```python
sigma = 1.1098
in_to_m = 0.0254

length_inches = 2.195
length_meters = 0.055753

mu_inches = np.log(length_inches)
mu_meters = np.log(length_meters)
mu_meters_fromInches = mu_inches + np.log(in_to_m)

lognorm_1 = lognorm(scale=np.exp(mu_meters), s=sigma)
lognorm_2 = lognorm(scale=np.exp(mu_meters_fromInches), s=sigma)
xspace = np.linspace(0, 1, 100)

fig, ax1 = plt.subplots(figsize=(4, 4))
ax1.plot(xspace, lognorm_1.cdf(xspace), 'k--', label='Dist. in Meters')
ax1.plot(xspace, lognorm_2.cdf(xspace), 'c:', label='Dist. in Inch \n Converted to Meters')
ax1.set_xlabel('Crack Length [m]')
ax1.set_ylabel('CDF')
plt.grid(color='gray', linestyle='--', alpha=0.3)
plt.legend(loc=0)

ax2 = ax1.twinx()
ax2.plot(xspace, lognorm_1.cdf(xspace)- lognorm_2.cdf(xspace), 'r-', alpha=0.4)
ax2.set_ylabel('Difference (CDF in meters - CDF in inches)')

plt.title('LogNormal Distribution w/ Unit Conversion');


fig, ax1 = plt.subplots(figsize=(4, 4))
ax1.plot(xspace, lognorm_1.pdf(xspace), 'k--', label='Dist. in Meters')
ax1.plot(xspace, lognorm_2.pdf(xspace), 'c:', label='Dist. in Meters \n Converted from Inches')
ax1.set_xlabel('Crack Length [m]')
ax1.set_ylabel('PDF')
plt.grid(color='gray', linestyle='--', alpha=0.3)
plt.legend(loc=0);
```

```python
sigma = 0.1

def c_to_k(temp):
    return temp + 273.15

def f_to_k(temp):
    return (temp - 32) * 5/9 + 273.15

T_C = 20
T_K = T_C + 273.15
T_F = 68
T_K = (T_F - 32)*5/9 + 273.15

mu_K = np.log(T_K)
mu_C = np.log(T_C)
mu_F = np.log(T_F)
mu_K_fromC = np.log( np.exp(mu_C) + 273.15 )
mu_K_fromF = np.log( (np.exp(mu_F) - 32)*5/9 + 273.15 )

lognorm_1 = lognorm(scale=np.exp(mu_K), s=sigma)
lognorm_2 = lognorm(scale=np.exp(mu_K_fromC), s=sigma)
lognorm_3 = lognorm(scale=np.exp(mu_K_fromF), s=sigma)
xspace = np.linspace(200, 400, 100)

fig, ax1 = plt.subplots(figsize=(4, 4))
ax1.plot(xspace, lognorm_1.cdf(xspace), 'k--', label='Dist. in K')
ax1.plot(xspace, lognorm_2.cdf(xspace), 'c:', label='Dist. in C \n Converted to K')
ax1.plot(xspace, lognorm_3.cdf(xspace), 'b-.', label='Dist. in F \n Converted to K')

ax1.set_xlabel('Temperature [K]')
ax1.set_ylabel('CDF')
plt.grid(color='gray', linestyle='--', alpha=0.3)
plt.legend(loc=0)

ax2 = ax1.twinx()
ax2.plot(xspace, lognorm_1.cdf(xspace)- lognorm_2.cdf(xspace), 'r-', alpha=0.4)
ax2.plot(xspace, lognorm_1.cdf(xspace)- lognorm_3.cdf(xspace), 'r--', alpha=0.4)
ax2.set_ylabel('Difference (CDF in K - CDF in C or F)')

plt.title('LogNormal Distribution w/ Unit Conversion')



fig, ax1 = plt.subplots(figsize=(4, 4))
ax1.plot(xspace, lognorm_1.pdf(xspace), 'k--', label='Dist. in K')
ax1.plot(xspace, lognorm_2.pdf(xspace), 'c:', label='Dist. in K \n Converted from C')
ax1.plot(xspace, lognorm_3.pdf(xspace), 'g-.', label='Dist. in K \n Converted from F')
ax1.set_xlabel('Temperature [K]')
ax1.set_ylabel('PDF')
plt.grid(color='gray', linestyle='--', alpha=0.3)
plt.legend(loc=0);
;
```
